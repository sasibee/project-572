```python
# Core
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Modeling
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix, RocCurveDisplay
)

# Models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.ensemble import RandomForestClassifier

# Reproducibility
SEED = 42
np.random.seed(SEED)
```

```python
df = pd.read_csv("titanic_augmented.csv")

df.head()
```

```python
df.info()
df.isna().sum()
```

```python
#feature selection
TARGET = "Survived"

drop_cols = [
    "PassengerId",
    "Name",
    "Ticket",
    "Cabin",
    "booking_reference",
    "service_id"
]

X = df.drop(columns=[TARGET] + drop_cols)
y = df[TARGET]

X.shape, y.shape
```

```python
#data splitting
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    stratify=y,
    random_state=SEED
)

print("Train size:", X_train.shape)
print("Test size:", X_test.shape)
print("Train survival rate:", y_train.mean())
print("Test survival rate:", y_test.mean())
```

```python
#feature typing
num_features = X_train.select_dtypes(include=["int64", "float64"]).columns.tolist()
cat_features = X_train.select_dtypes(include=["object"]).columns.tolist()

print("Numeric features:", num_features)
print("Categorical features:", cat_features)
```

```python
#data preprocessing
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_features),
        ("cat", categorical_transformer, cat_features),
    ],
    remainder="drop"
)

preprocess_scaled = preprocessor
```

```python
# --- No-scale preprocessor for tree-based models (RF/DT) ---

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

numeric_noscale = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

preprocess_noscale = ColumnTransformer(
    transformers=[
        ("num", numeric_noscale, num_features),
        ("cat", categorical_transformer, cat_features),  # <- use your existing transformer
    ],
    remainder="drop"
)
```

```python
from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, roc_auc_score, log_loss
import numpy as np

def _proba_or_none(model, X):
    if hasattr(model, "predict_proba"):
        return model.predict_proba(X)[:, 1]
    return None

def fit_tune_evaluate(
    model_name,
    clf,
    preprocess,
    param_grid,
    X_train,
    y_train,
    X_test,
    y_test,
    scoring="accuracy"
):
    pipe = Pipeline([
        ("preprocess", preprocess),
        ("clf", clf)
    ])

    grid = GridSearchCV(
        pipe,
        param_grid=param_grid,
        scoring=scoring,
        cv=cv,
        n_jobs=-1,
        refit=True
    )
    grid.fit(X_train, y_train)

    best_model = grid.best_estimator_

    y_pred = best_model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)

    y_proba = _proba_or_none(best_model, X_test)
    auc = roc_auc_score(y_test, y_proba) if y_proba is not None else np.nan
    ll  = log_loss(y_test, y_proba) if y_proba is not None else np.nan

    try:
        n_features = len(
            best_model.named_steps["preprocess"].get_feature_names_out()
        )
    except:
        n_features = np.nan

    return {
        "model": model_name,
        "best_params": grid.best_params_,
        "cv_best_score": grid.best_score_,
        "test_accuracy": acc,
        "test_auc": auc,
        "test_logloss": ll,
        "n_features": n_features,
        "estimator": best_model
    }
```

```python
results = []
```

### Model A：Logistic Regression（using scaled preprocess）

```python
from sklearn.linear_model import LogisticRegression

log_clf = LogisticRegression(max_iter=2000, random_state=SEED)

log_param_grid = {
    "clf__C": [0.01, 0.1, 1, 10],
    "clf__penalty": ["l2"],
    "clf__solver": ["lbfgs"]
}

log_result = fit_tune_evaluate(
    model_name="LogisticRegression",
    clf=log_clf,
    preprocess=preprocess_scaled,
    param_grid=log_param_grid,
    X_train=X_train,
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    scoring="accuracy"
)

results.append(log_result)
log_result["best_params"], log_result["cv_best_score"], log_result["test_accuracy"]
```

### Model B：KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn_clf = KNeighborsClassifier()

knn_param_grid = {
    "clf__n_neighbors": [3, 5, 7, 9, 11, 15],
    "clf__weights": ["uniform", "distance"],
    "clf__p": [1, 2]
}

knn_result = fit_tune_evaluate(
    model_name="KNN",
    clf=knn_clf,
    preprocess=preprocess_scaled,
    param_grid=knn_param_grid,
    X_train=X_train,
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    scoring="accuracy"
)

results.append(knn_result)
knn_result["best_params"], knn_result["cv_best_score"], knn_result["test_accuracy"]
```

### Model C：Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier(
    random_state=SEED,
    n_jobs=-1
)

rf_param_grid = {
    "clf__n_estimators": [200, 400],
    "clf__max_depth": [None, 6, 10],
    "clf__min_samples_leaf": [1, 2, 4]
}

rf_result = fit_tune_evaluate(
    model_name="RandomForest",
    clf=rf_clf,
    preprocess=preprocess_noscale,
    param_grid=rf_param_grid,
    X_train=X_train,
    y_train=y_train,
    X_test=X_test,
    y_test=y_test,
    scoring="accuracy"
)

results.append(rf_result)
rf_result["best_params"], rf_result["cv_best_score"], rf_result["test_accuracy"]
```
